```python
for p in [0.15, 0.3,0.5,0.7, 0.85]:
	q = 1-p
	number = 0
	for i in range(1000):
		z=1
		for _ in range(500):
			z = st.geom.rvs(p, size = z, loc = -1).sum() # loc = -1 - считаем только потомков
			if z == 0:
				number += 1
				break
			if z > 5000:
				break
	print(f"p = {p} => оценка = {number/1000}")
	print(number)
	print("точное значение:", np.minimum(p/q, 1))
	print('\n')
    
```
